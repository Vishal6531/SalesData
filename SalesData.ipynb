```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the sales data
data = pd.read_csv('new_sales_data_50.csv')
```

```python
# Explore sales data
print(data.info())
```

```python
# Data Analysis
# Calculate total revenue and total units sold for each product
product_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
product_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
print(product_quantity)
print(product_revenue)
```

```python
# Analyze sales trends over time (daily revenue trend)
daily_revenue = data.groupby('Date')['Revenue'].sum()
print(daily_revenue.head())
```

```python
# Identify the best-selling product categories
category_revenue = data.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
print(category_revenue)
```

```python
# Plotting visualizations
# 1. Bar chart displaying total sales for each product
plt.figure(figsize=(10, 5))
plt.bar(product_revenue.index, product_revenue.values, color='green')
plt.title('Total Revenue by Product', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Bar chart displaying total quantity sold for each product
plt.figure(figsize=(10, 5))
plt.bar(product_quantity.index, product_quantity.values, color='orange')
plt.title('Total Units Sold by Product', fontsize=14)
plt.xlabel('Product', fontsize=12)
plt.ylabel('Units Sold', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 1. Line graph showing the trend of total revenue over time
plt.figure(figsize=(10, 5))
plt.plot(daily_revenue.index, daily_revenue.values, marker='o', color='b')
plt.title('Daily Revenue Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 4. Pie chart showing the percentage of revenue by category
plt.figure(figsize=(6, 6))
plt.pie(category_revenue.values, labels=category_revenue.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Revenue Percentage by Category', fontsize=14)
plt.tight_layout()
plt.show()
```
